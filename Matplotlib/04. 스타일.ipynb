# 4. 스타일


```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python

x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,linewidth = 5)
```

## Marker


```python
plt.plot(x,y,marker="o")
```

```python
plt.plot(x,y,marker="o", linestyle = 'None')
```

```python
plt.plot(x,y,marker = "v")
```

```python
plt.plot(x,y,marker = "v", markersize = 10)
```

```python
plt.plot(x,y,marker = "X")
```

```python
plt.plot(x,y,marker = "v", markeredgecolor = 'red')
```

```python
plt.plot(x,y,marker = "o", markeredgecolor = 'red')
```

```python
plt.plot(x,y,marker = "o",markersize =20 ,markeredgecolor = 'red', markerfacecolor = 'yellow')
```

## 선 스타일


```python
plt.plot(x,y, linestyle = ':')
```

```python
plt.plot(x,y, linestyle = '--')
```

```python
plt.plot(x,y, linestyle = '-.')
```

```python
plt.plot(x,y, linestyle = '-')
```

## 색깔

```python
plt.plot(x,y,color = 'pink')
```

```python
plt.plot(x,y,color = '#ff0000')
```

```python
plt.plot(x,y,color = 'b')
```

```python
plt.plot(x,y,color = 'g')
```

## 포맷

```python
plt.plot(x,y,'ro--') # color , marker, linestyle
```

```python
plt.plot(x,y,'bv:')
```

```python
plt.plot(x,y,'go')
```

```python
plt.plot(x,y,color='g',marker='o',linestyle='None')
```

## 축약어


```python
plt.plot(x,y,marker='o', mfc = 'red', ms = 10, mec = 'blue', ls = ':')
```

## 투명도

```python
plt.plot(x,y,marker='o', mfc = 'red', ms = 10, alpha = 0.4)
```

## 그래프 크기

```python
plt.figure(figsize = (10,5))
plt.plot(x,y)
```

```python
plt.figure(figsize = (5,10))
plt.plot(x,y)
```

```python
plt.figure(figsize=(10,5), dpi = 200) # 해상도 dots per inch (확대의 의미로 이해하면 될듯)
plt.plot(x,y)
```

## 배경색


```python
plt.figure(facecolor = 'yellow')
plt.plot(x,y)
```

```python
plt.figure(facecolor = '#a1c3ff')
plt.plot(x,y)
```
