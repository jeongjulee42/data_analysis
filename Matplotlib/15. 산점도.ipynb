# 15. 산점도 그래프


```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic'
# matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
df = pd.read_excel('../Pandas/score.xlsx')
df
```

```python
df['학년'] = [3,3,2,1,1,3,2,2]
df
```

```python
plt.scatter(df['영어'],df['수학'])
plt.xlabel('영어점수')
plt.ylabel('수학점수')
```

```python
import numpy as np
sizes = np.random.rand(8) * 1000
sizes
```

```python
plt.scatter(df['영어'],df['수학'], s = sizes)
plt.xlabel('영어점수')
plt.ylabel('수학점수')
```

```python
sizes = df['학년'] * 500 # 학년별로 크기 다르게

plt.scatter(df['영어'],df['수학'], s = sizes)
plt.xlabel('영어점수')
plt.ylabel('수학점수')
```

```python
plt.scatter(df['영어'],df['수학'], s = sizes, c=df['학년'], cmap = 'viridis') # 색상 알아서 cmap에서 제공해줌
plt.xlabel('영어점수')
plt.ylabel('수학점수')
```

```python
plt.scatter(df['영어'],df['수학'], s = sizes, c=df['학년'], cmap = 'viridis', alpha = 0.3) # 색상 알아서 cmap에서 제공해줌
plt.xlabel('영어점수')
plt.ylabel('수학점수')
```

```python
plt.figure(figsize=(7,7))

plt.scatter(df['영어'],df['수학'], s = sizes, c=df['학년'], cmap = 'viridis', alpha = 0.3) 
plt.xlabel('영어점수')
plt.ylabel('수학점수')
plt.colorbar()
```

```python
plt.figure(figsize=(7,7))

plt.scatter(df['영어'],df['수학'], s = sizes, c=df['학년'], cmap = 'viridis', alpha = 0.3) 
plt.xlabel('영어점수')
plt.ylabel('수학점수')
plt.colorbar(ticks =[1,2,3], label='학년', shrink = 0.5) #shrink - 컬러바 줄이기
```

```python
plt.figure(figsize=(7,7))

plt.scatter(df['영어'],df['수학'], s = sizes, c=df['학년'], cmap = 'viridis', alpha = 0.3) 
plt.xlabel('영어점수')
plt.ylabel('수학점수')
plt.colorbar(ticks =[1,2,3], label='학년', shrink = 0.5, orientation = 'horizontal')
```
