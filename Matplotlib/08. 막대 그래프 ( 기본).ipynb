# 8. 막대 그래프 (기본)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
labels = ['강백호','서태웅','정대만']
values = [190,187,184]

plt.bar(labels, values)
```

```python
labels = ['강백호','서태웅','정대만']
values = [190,187,184]

plt.bar(labels, values, color = 'r')
```

```python
labels = ['강백호','서태웅','정대만']
values = [190,187,184]
colors = ['r','g','b']

plt.bar(labels, values, color = colors)
```

```python
labels = ['강백호','서태웅','정대만']
values = [190,187,184]
colors = ['r','g','b']

plt.bar(labels, values, color = colors, alpha = 0.5)
```

```python
labels = ['강백호','서태웅','정대만']
values = [190,187,184]


plt.bar(labels, values)
plt.ylim(175,195)
```

```python
plt.bar(labels, values, width = 0.5)
```

```python
plt.bar(labels, values, width = 0.3)
plt.xticks(rotation = 45) # x축의 이름 데이터 각도를 45도로 설정
plt.yticks(rotation = 45)
```

```python
labels = ['강백호','서태웅','정대만']
values = [190,187,184]
ticks = ['1번','2번','3번']

plt.bar(labels, values)
plt.xticks(labels,ticks, rotation = 90)
```
