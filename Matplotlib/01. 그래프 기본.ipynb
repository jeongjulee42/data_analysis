# Matplotlib
다양한 형태의 그래프를 통해 데이터 시각화를 할수있는 라이브러리


```python
import matplotlib.pyplot as plt
```

## 1. 그래프 기본

```python
x = [1,2,3]
y = [2,4,8]
plt.plot(x, y)
```

```python
print(plt.plot(x, y))
```

```python
plt.plot(x, y)
plt.show()
```

## Title 설정

```python
plt.plot(x,y)
plt.title('Line Graph')
```

```python
plt.plot(x,y)
plt.title('꺾은선 그래프')
# 한글을 쓰려면 추가 환경설정 필요
```

## 한글 폰트 설정

```python
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['font.size'] = 15 # 글자 크기
```

```python
plt.plot(x,y)
plt.title('꺾은선 그래프')
```

```python
import matplotlib.font_manager as fm
fm.fontManager.ttflist # 사용 가능한 폰트 확인
```
