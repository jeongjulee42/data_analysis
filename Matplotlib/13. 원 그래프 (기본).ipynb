# 13. 원 그래프 (기본)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic'
# matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
```

```python
values = [30,25,20,13,10,2]
labels = ['Python','Java','Js','c#','c/c++','etc']

plt.pie(values)
plt.show()
```

```python
plt.pie(values, labels = labels)
plt.show()
```

```python
plt.pie(values, labels = labels , autopct = '%.1f')
plt.show()
```

```python
plt.pie(values, labels = labels , autopct = '%.1f%%')
plt.show()
```

```python
plt.pie(values, labels = labels , autopct = '%.1f%%', startangle = 90)
plt.show()
```

```python
plt.pie(values, labels = labels , autopct = '%.1f%%', startangle = 90, counterclock = False)
plt.show()
```

```python
# explode = [0.2,0.1,0,0,0,0] # 띄울 간격

explode = [0.05] * 6
plt.pie(values, labels = labels , explode = explode)

plt.show()
```

```python
plt.pie(values, labels = labels , explode = explode)
plt.legend()
plt.show()
```

```python
plt.pie(values, labels = labels , explode = explode)
plt.legend(loc = (1.2,0.3))
plt.show()
```

```python
plt.pie(values, labels = labels , explode = explode)
plt.legend(loc = (1.2,0.3), title = '언어별 선호도')
plt.show()
```

```python
plt.pie(values, labels = labels , explode = explode)
plt.title('언어별 선호도')
plt.legend(loc = (1.2,0.3), title = '언어별 선호도')
plt.show()
```
